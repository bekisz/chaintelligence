# USD M2 Money Supply

Downloded from [FED Site here](https://fred.stlouisfed.org/series/WM2NS)

```python
import pandas as pd
df = pd.read_csv('data/WM2NS.csv')
df['DATE']= pd.to_datetime(df['DATE'])
df.set_index(['DATE'],inplace=True, verify_integrity = True, drop = True)
```

```python
import matplotlib.pyplot as plt

plt.title("USD M2 Money Supply", fontsize=20)
plt.ylabel("Billion USD")
plt.xlabel('Date')

F = plt.gcf()
DefaultSize = F.get_size_inches()
df['WM2NS'].plot(c='blue',label='M2',linewidth=2)
plt.grid()
plt.ylim((0,25000))
F.set_size_inches( (DefaultSize[0]*2, DefaultSize[1]*1) )

```
